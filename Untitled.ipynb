```python
import pandas as pd
import numpy as np 
import re, json, os , logging , random, html, datetime, openpyxl 
from pdfminer.high_level import extract_text
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer
```

```python
def trans_normed(t):
    if "YELLOW CAB" in t:
        return "CAB"
    if re.findall("CONOCO|RENO|ALBERT", t ) :
        return "FOOD"
    if re.findall("SPECT|COOPER|PROG|NORTHWEST|MISSOU|REPUBLIC", t.upper() ) :
        return "UTIL"
    if "PAYROLL" in t:
        return "PAYROLL" 
    return t
df=pd.read_csv(r'C:\Users\timko\Documents\LEE\Checking1.csv')
df['trans_normed'] = df['trans'].apply(trans_normed)
df['m'] = df['date'].apply(lambda s: re.sub("[^0-9]",  "", s[:2]) )
df.to_csv(r'C:\Users\timko\Documents\LEE\Checking2.csv', index=False)
df.groupby(['m','trans_normed'], as_index=False).sum()
```
