```python
engine = create_engine(connstr) 
def get_df(query):
    conn = engine.connect() 
    df = pd.read_sql(query,con=conn) 
    conn.close()
    return df
```

```python
def normalizedf(df):
    df.drop( columns=[c for c in df.columns if '__' in c ] , inplace=True )   
    cols = list(df)   
    df[cols] = df[cols].astype(str)
    df['IDT'] = df['IDTEXT'].apply(id_to_var) 
    return df
```

```python
def id_split(s):
    s = s + '000'
    s = re.sub('[^\d]', '', s)  
    return s[:4] 
print( id_split('5.1-4') )
def id_to_var(s): 
    s = re.sub('\s|\.', '_', s)  
    return s 
print( id_to_var('5.1 4') )
```
