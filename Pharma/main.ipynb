```python
import pandas as pd
import pymysql
import sqlalchemy
from sqlalchemy import create_engine
import pgeocode
import re
import random
import requests
import json
from random import randrange  
from barnum import gen_data
```

```python
genders = ['Male','Female']
eth_list = ['Cauc','Latino', 'Asian', 'AM']
mstat_list = ['S','M','D'] 
dl=[]
dfp = pd.read_csv('data/Patient_metadata.csv') 
for i,r in dfp.iterrows():
    gen = genders[randrange(len(genders))] 
    eth = eth_list[randrange(len(eth_list))] 
    mstat = mstat_list[randrange(len(mstat_list))] 
    zp, city, state = gen_data.create_city_state_zip()
    fname, lname = gen_data.create_name(gender=f'{gen}') 
    dl.append({
          'row':str(i)
        , 'Patient_ID':r['Patient_ID']
        , 'DrugRegimen':r['DrugRegimen']
        , 'Sex':r['Sex']
        , 'Age_months':r['Age_months']
        , 'Weight':r['Weight']
        , 'Name': fname + ' P. ' +lname        
        , 'Marital Status': mstat
        , 'Ethnicity': eth       
        , 'Address': '123 Elm ST. ' +  city + ' ' + state + ' ' + zp + ' '
        , 'ContactNumbers':  'p: ' + str(r['Weight']) + '1-12'+ str(r['Weight']) + ', c: 3' + str(r['Weight']) + '-3412'+ ', fax: 9' + str(r['Weight']) + '-5555'  
    })  
df = pd.DataFrame(dl)     
df.to_csv('data/Patient_meta.csv', index=False)
df
```
