```python
def clean_text(txt):
    txt=re.sub('Chapter.*\d{1,2}|,|’|“|”','',txt)
    txt=' '.join([w for w in txt.lower().split(' ') if not w in sw])
    return txt
txt = ''
with open(r'C:\Users\timko\Downloads\sjm.txt', 'r') as f:
    txt=clean_text(f.read()) 
sent = sent_tokenize(txt)
```

```python
txt1 = sent
tf = TfidfVectorizer(smooth_idf=False, sublinear_tf=False, norm=None, analyzer='word')
txt_fitted = tf.fit(txt1)
txt_transformed = txt_fitted.transform(txt1)
print ( txt1)
```
